```python
from keybert import KeyBERT
import pandas as pd

a = pd.read_csv('../trip_advisor/crawling/크롤러 데이터/성산일출봉_reviews.csv', encoding='utf-8')
review = a.values.tolist()
```

```python
kw_model = KeyBERT()
keywords = keywords = kw_model.extract_keywords(review,keyphrase_ngram_range=(2,4),use_maxsum = True,top_n = 20)
```

```python
keywords_mmr = kw_model.extract_keywords(review,keyphrase_ngram_range=(2,4),use_mmr = True,top_n = 20,diversity = 0.3)
keywords_mmr
```
