```python
b = pd.read_csv('place_data.csv').drop('관광지 이름', axis = 1)
b
```

```python
import pandas as pd
import re

# 예시 데이터프레임 생성
data = b
df = pd.DataFrame(data)

# 정규 표현식을 사용하여 (우)와 숫자를 포함한 부분을 제거
df['place_location'] = df['place_location'].apply(lambda x: re.sub(r'\(우\)\d+', '', x))

df
```

```python
df.to_csv("place_data.csv", mode='w')
```
