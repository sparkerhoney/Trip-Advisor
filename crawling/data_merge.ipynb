```python
import pandas as pd

place_data = pd.read_csv('./place_data.csv')
place_topic = pd.read_csv('./place_topic.csv')
```

```python
place_data.head()
```

```python
place_topic = place_topic.replace('로맨틱', '낭만적인')
```

```python
place_data.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
merge_df = pd.merge(place_data, place_topic, on = 'place_name')
merge_df.drop_duplicates(subset='place_name', inplace=True)
```

```python
merge_df.to_excel('place_fin.xlsx', index=False)
```
