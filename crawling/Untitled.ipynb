```python
import pandas as pd

# 엑셀 파일에서 시트1을 데이터프레임으로 읽어오기
df = pd.read_excel('place_fin.xlsx', sheet_name='Sheet1')
```

```python
df
```

```python
keyword_df = df[df[['topic1', 'topic2', 'topic3']].apply(lambda row: any(keyword in row.values for keyword in ['감성적인', '낭만적인', '분위기']), axis=1)]
keyword_df
```

```python
keyword_df.to_excel('감성적인, 낭만적인, 분위기.xlsx', index=False)
```
